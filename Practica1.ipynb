### 1. Crea un circuito cuántico con un solo qubit y aplícale 5 compuertas diferentes:

Primero crearemos el qubit:

```python
from qiskit import QuantumRegister
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex

qubitEj1 = QuantumRegister(1, 'q1')
print(qubitEj1)
```

A continuación creamos el circuito y lo visualizamos:

```python
circuito = QuantumCircuit(qubitEj1)
psi_0 = Statevector(circuito) #Se guarda el vector de estado despues de crear el qubit
circuito.draw('mpl')
```

**A. Guarda el vector de estado justo después de la creación del qubit y después de cada compuerta aplicada, muestra esos vectores de estado (en total serán 6).**

```python
print("Vector de estado del qubit original: ")
print(display(array_to_latex(psi_0.data)))
```

```python
circuito.x(0)                # Se aplicó la compuerta X (NOT) al qubit
psi_1 = Statevector(circuito)  #Se guarda el vector de estado
print("Vector de estado del qubit despues de aplicarle la compuerta X: ")
print(display(array_to_latex(psi_1.data)))

circuito.draw('mpl')         # mostrar el circuito
```

```python
circuito.y(0)                # Se aplicó la compuerta Y al qubit
psi_2 = Statevector(circuito)  #Se guarda el vector de estado
print("Vector de estado del qubit despues de aplicarle la compuerta Y: ")
print(display(array_to_latex(psi_2.data)))

circuito.draw('mpl')         # mostrar el circuito
```

```python
circuito.z(0)                # Se aplicó la compuerta z al qubit
psi_3 = Statevector(circuito)  #Se guarda el vector de estado
print("Vector de estado del qubit despues de aplicarle la compuerta Z: ")
print(display(array_to_latex(psi_3.data)))

circuito.draw('mpl')         # mostrar el circuito
```

```python
circuito.h(0)                # Se aplicó la compuerta H al qubit
psi_4 = Statevector(circuito)  #Se guarda el vector de estado
print("Vector de estado del qubit despues de aplicarle la compuerta H: ")
print(display(array_to_latex(psi_4.data)))

circuito.draw('mpl')         # mostrar el circuito
```

```python
circuito.s(0)                # Se aplicó la compuerta S al qubit
psi_5= Statevector(circuito)  #Se guarda el vector de estado
print("Vector de estado del qubit despues de aplicarle la compuerta S: ")
print(display(array_to_latex(psi_5.data)))

circuito.draw('mpl')         # mostrar el circuito
```

**B. En una celda de markdown explica que es lo que le sucede al qubit conforme se le va aplicando cada compuerta.**

**Compuerta X** 

El valor de su vector de estado cambia, su valor inicial era de **[1 0]** o y aplicando la compuerta intercambia los estados, obteniendo un valor de **[0 1]**

**Compuertas XY** 

El valor de su vector de estado cambia ya que realiza una rotación del estado con una fase, su valor inicial era de **[0 1]** y aplicando la compuerta se obtiene un valor de **[-i 0]**

**Compuertas XYZ**

El valor de su vector de estado no cambia debido a que la compuerta Z sólo introduse el cambio de fase en el estado **[-i 0]** por lo que el valor se mantiene en **[-i 0]**

**Compuertas XYZH**

El valor de su vector de estado cambia ya que coloca un qubit en una superposición uniforme, su valor inicial era de **[-i 0]** y aplicando la compuerta intercambia los estados, obteniendo un valor de **[$-\frac{\sqrt{2}i}{2}$ $-\frac{\sqrt{2}i}{2}$]**

**Compuertas XYZHS** 

El valor de su vector de estado cambia ya que se introduce una fase de 90° en el estado **[0 1]**, su valor inicial era de **[$-\frac{\sqrt{2}i}{2}$ $-\frac{\sqrt{2}i}{2}$]** y aplicando la compuerta cambia el signo del estado correspondiente debido a la fase, obteniendo un valor de **[$-\frac{\sqrt{2}i}{2}$ $\frac{\sqrt{2}}{2}$]**

**C. Escribe con LATEX la expresión matemática completa de la aplicación de las 5 compuertas (no olvides poner atención en el orden). Revisa la segunda celda de este notebook, ¿hay algo útil ahí?.**

Tenemos nuestro qubit:
$$
| q_1 \rangle = | 0 \rangle
$$

Regresentado como vector:
$$
q_1 = \begin{pmatrix}
1 \\
0
\end{pmatrix}
$$

**Ahora, aplicamos la compuerta X al qubit:**
$$
X \cdot q_1 = \begin{pmatrix}
0 & 1 \\
1 & 0
\end{pmatrix} \cdot \begin{pmatrix}
1 \\
0
\end{pmatrix}
$$

Realizamos la multiplicación de matrices:
$$
X \cdot q_1 = \begin{pmatrix}
0 \cdot 1 + 1 \cdot 0 \\
1 \cdot 1 + 0 \cdot 0
\end{pmatrix} = \begin{pmatrix}
0 \\
1
\end{pmatrix}
$$

**Después de aplicar la compuerta X, finalmente obtenemos:**
$$
q_1' = \begin{pmatrix}
0 \\
1
\end{pmatrix}
$$

**Ahora aplicamos la compuerta Y a la matriz resultante q1':**
$$
Y \cdot q_1' = \begin{pmatrix}
0 & -i \\
i & 0
\end{pmatrix} \cdot \begin{pmatrix}
0 \\
1
\end{pmatrix}
$$

Realizamos la multiplicación de matrices:
$$
Y \cdot q_1' = \begin{pmatrix}
0 \cdot 0 + (-i) \cdot 1 \\
i \cdot 0 + 0 \cdot 1
\end{pmatrix} = \begin{pmatrix}
-i \\
0
\end{pmatrix}
$$

**Después de aplicar la compuerta Y, finalmente obtenemos:**
$$
q_1'' = \begin{pmatrix}
-i \\
0
\end{pmatrix}
$$

**Ahora aplicamos la compuerta Z a la matriz resultante q1'':**
$$
Z \cdot q_1'' = \begin{pmatrix}
1 & 0 \\
0 & -1
\end{pmatrix} \cdot \begin{pmatrix}
-i \\
0
\end{pmatrix}
$$

Realizamos la multiplicación de matrices:
$$
Z \cdot q_1'' = \begin{pmatrix}
1 \cdot (-i) + 0 \cdot 0 \\
0 \cdot (-i) + (-1) \cdot 0
\end{pmatrix} = \begin{pmatrix}
-i \\
0
\end{pmatrix}
$$

**Después de aplicar la compuerta Z, finalmente obtenemos:**
$$
q_1''' = \begin{pmatrix}
-i \\
0
\end{pmatrix}
$$

**Ahora aplicamos la compuerta H a la matriz resultante q1''':**
$$
H \cdot q_1''' = \frac{1}{\sqrt{2}} \begin{pmatrix}
1 & 1 \\
1 & -1
\end{pmatrix} \cdot \begin{pmatrix}
-i \\
0
\end{pmatrix}
$$

Realizamos la multiplicación de matrices:
$$
H \cdot q_1''' = \frac{1}{\sqrt{2}} \begin{pmatrix}
-i \\
-i
\end{pmatrix} = \begin{pmatrix}
\frac{-i}{\sqrt{2}} \\
\frac{-i}{\sqrt{2}}
\end{pmatrix}
$$

**Después de aplicar la compuerta H, finalmente obtenemos:**
$$
q_1'''' = \begin{pmatrix}
\frac{-i}{\sqrt{2}} \\
\frac{-i}{\sqrt{2}}
\end{pmatrix}
$$

**Ahora aplicamos la compuerta S a la matriz resultante q1'''':**
$$
S \cdot q_1'''' = \begin{pmatrix}
1 & 0 \\
0 & i
\end{pmatrix} \cdot \begin{pmatrix}
\frac{-i}{\sqrt{2}} \\
\frac{-i}{\sqrt{2}}
\end{pmatrix}
$$

Realizamos la multiplicación de matrices:
$$
S \cdot q_1'''' = \begin{pmatrix}
1 \cdot \frac{-i}{\sqrt{2}} + 0 \cdot \frac{-i}{\sqrt{2}} \\
0 \cdot \frac{-i}{\sqrt{2}} + i \cdot \frac{-i}{\sqrt{2}}
\end{pmatrix} = \begin{pmatrix}
\frac{-i}{\sqrt{2}} \\
\frac{1}{\sqrt{2}}
\end{pmatrix}
$$

**Después de aplicar la compuerta S, finalmente obtenemos:**
$$
q_1''''' = \begin{pmatrix}
\frac{-i}{\sqrt{2}} \\
\frac{1}{\sqrt{2}}
\end{pmatrix}
$$
